```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
```

```python
#review > div.offer-review__list > div:nth-child(1) > div:nth-child(45) > div.offer-review__list--wrapper > div > div
ban_a = [22031, 6242, 22027, 19879] # 시장, 쿠킹클래스
new_a = [5665, 2097, 9660, 5250] # 야경, 미술
ha_a = [32529, 6507, 41529, 7443] #스노쿨링, 목장, 스카이다이빙
activity = [ban_a, new_a, ha_a]
```

```python
temp = []

for acts in activity:
    for act in acts:
        driver = webdriver.Chrome()
        driver.get("https://www.myrealtrip.com/offers/{}".format(act))
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        time.sleep(3)
        for num in range(1, 25):
            element = driver.find_element_by_css_selector('#review > div.offer-review__list > div.offer-review--more > button')
            driver.execute_script("arguments[0].click();", element)
            print(num)
            time.sleep(1)

        for i in range(1, 60):
            try:
                a = driver.find_element_by_css_selector('#review > div.offer-review__list > div:nth-child(1) > div:nth-child({})'.format(i)).text
                temp.append({'나라':acts, 'activity' : act, '리뷰' : a})
            except:
                pass
            
            time.sleep(2)
        driver.quit()
```

```python
t = pd.DataFrame(temp)
t['나라'] = t['나라'].astype(str)
t.drop_duplicates()
```

```python
t['나라'][t['나라'] == '[22031, 6242, 22027, 19879]'] = '방콕'
t['나라'][t['나라'] == '[5665, 2097, 9660, 5250]'] = '뉴욕'
t['나라'][t['나라'] == '[32529, 6507, 41529, 7443]'] = '하와이'
t
```
